```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 1. Asegurarnos de que existe la carpeta para guardar resultados
os.makedirs('results', exist_ok=True)

# 2. Cargar datos
csv_path = 'Anotaciones/LUNA25_Public_Training_Development_Data.csv'
df = pd.read_csv(csv_path)

# 3. Graficar
plt.figure(figsize=(10, 6))
# Usamos hue='label' para evitar el Warning y que se vea pro
sns.countplot(x='label', data=df, hue='label', palette='viridis', legend=False)

plt.title('Distribución de Niveles de Malignidad - LUNA25', fontsize=15)
plt.xlabel('Nivel de Malignidad (1-5)', fontsize=12)
plt.ylabel('Cantidad de Nódulos', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)

# 4. Guardar y mostrar
plt.savefig('results/distribucion_malignidad.png')
plt.show()

# Ver los números exactos
print("\nConteo por nivel:")
print(df['label'].value_counts().sort_index())
```
